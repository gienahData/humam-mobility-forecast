### Load packages

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import json
import pandas as pd
```

### Find files

```python
!ls /media/Data_storage/Mobilcell/DayGraphAnalytics | tail
```

```python
source = '/media/Data_storage/Mobilcell/DayGraphAnalytics_unweighted/'
#source = '/media/Data_storage/Mobilcell/DayGraphAnalytics_unweighted/'
files_directed = np.array( sorted([ i for i in os.listdir(source) if 'global_attributes_directed' in i ]) )
files_undirected = np.array( sorted([ i for i in os.listdir(source) if 'global_attributes_undirected' in i ]) )
dates = np.array( [ i.split('_')[-1][:-5] for i in files_undirected ] )
files_directed.shape, files_undirected.shape, files_directed[:5], files_undirected[:5], dates[:5]
```

#### Open first and save header

how to flatten json: https://mindtrove.info/flatten-nested-json-with-pandas/

```python
with open(source+files_undirected[0]) as json_file:
    data = json.load(json_file)
    msgs = pd.json_normalize( data )
    msgs.insert(0, 'date', dates[0])
    header = list(msgs) # [ i.replace('.0', '-0') for i in list(msgs) ]
```

```python
header
```

```python
def do_work( files ):
    df_rows = []
    #print(files)
    for n in range(files.shape[0]):
        with open(source+files[n]) as json_file:
            data = json.load(json_file)
            msgs = pd.json_normalize( data )
            msgs.insert(0, 'date', dates[n])
            #print( files[n], msgs.values.shape )
            df_rows.append( msgs.values )
    return np.concatenate(df_rows, axis=0)
```

```python
ret = do_work(files_undirected)
save_df = pd.DataFrame( ret, columns=header )
save_df.shape
```

```python
save_df.head()
```

#### Save to disk

```python
save_df.to_csv( source+'graph_global_attributes_undirected_unweighted.csv', index=False )
```

```python
ret = do_work(files_directed)
save_df = pd.DataFrame( ret, columns=header )
save_df.shape
```
